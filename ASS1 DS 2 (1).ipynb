```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from sklearn.feature_selection import chi2

# Load the dataset
file_path = 'CrimesOnWomenData.csv'  # Path to the uploaded file
data = pd.read_csv(file_path)

# Display dataset information
print("Dataset Overview:")
print(data.info())
print("\nFirst 5 rows of the dataset:")
print(data.head())
```

WHATS DONE AND ACHIEVED BELOW

6. Visualizing Descriptive Statistics

```python
# Drop unnamed or redundant columns if applicable
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]

# Descriptive Statistics
print("\nDescriptive Statistics:")
descriptive_stats = data.describe(include='all').transpose()
modes = data.mode().iloc[0]
descriptive_stats['Mode'] = modes
print(descriptive_stats)

# Visualize Descriptive Statistics
descriptive_stats_numeric = data.describe()
descriptive_stats_numeric.plot(kind='bar', figsize=(15, 6))
plt.title('Descriptive Statistics of Numeric Features')
plt.xticks(rotation=45)
plt.show()
```

WHATS DONE AND ACHIEVED BELOW
8. Segregating Target and Explanatory Variables
Action: Identify the target variable(s) (e.g., Rape, DV) and explanatory variables by dropping unrelated columns.
Purpose: Sets up the data for correlation analysis and chi-square tests.
Importance: Helps focus analysis on the relationship between features and the target variables.

```python
# Segregate target variable and explanatory variables
# Assuming 'Rape' is one target variable and 'DV' is another if present
target_variable_rape = data['Rape'] if 'Rape' in data.columns else None
target_variable_dv = data['DV'] if 'DV' in data.columns else None
explanatory_variables = data.drop(columns=['Rape', 'DV', 'State', 'Year'], errors='ignore')
```

WHATS DONE AND ACHIEVED BELOW
9. Correlation Analysis
Action: Calculate and display correlation between the target variable and the explanatory variables.
Purpose: Shows which features have a strong linear relationship with the target, which is valuable for predictive modeling.
Importance: Correlation helps identify important variables and those that may not contribute much to model performance.

```python
# Correlation Analysis
print("\nCorrelation Analysis:")
if target_variable_rape is not None:
    correlation_results_rape = explanatory_variables.corrwith(target_variable_rape)
    print("Correlation with Rape:")
    print(correlation_results_rape)
    sns.barplot(x=correlation_results_rape.index, y=correlation_results_rape.values)
    plt.title('Correlation with Rape')
    plt.xticks(rotation=45)
    plt.show()

if target_variable_dv is not None:
    correlation_results_dv = explanatory_variables.corrwith(target_variable_dv)
    print("\nCorrelation with DV:")
    print(correlation_results_dv)
    sns.barplot(x=correlation_results_dv.index, y=correlation_results_dv.values)
    plt.title('Correlation with DV')
    plt.xticks(rotation=45)
    plt.show()
```

WHATS DONE AND ACHIEVED BELOW
10. Chi-Square Test for Independence
Action: Perform chi-square tests to assess if categorical explanatory variables are independent of the target variable.
Purpose: Determines if there is a statistically significant association between features and the target.
Importance: Useful for feature selection when dealing with categorical data to identify relevant features.
11. Converting Chi-Square Results to a DataFrame
Action: Format the chi-square test results into a structured DataFrame for easy viewing and analysis.
Purpose: Makes it easier to inspect and compare the results across different features.
Importance: Streamlines the process of identifying important features for further analysis.
12. Visualizing Chi-Square Test Results
Action: Create bar charts or other visualizations of chi-square values for features.
Purpose: Highlights which features show a stronger dependency on the target variable.
Importance: Visualization helps to quickly identify and focus on significant features.


```python
# Chi-square Test for Independence
chi_square_results = {}
for column in explanatory_variables.columns:
    if target_variable_rape is not None:
        contingency_table_rape = pd.crosstab(target_variable_rape, explanatory_variables[column])
        chi2_rape, p_rape, _, _ = chi2_contingency(contingency_table_rape)
        chi_square_results[f"{column}_Rape"] = {"Chi2 Value": chi2_rape, "P-Value": p_rape}

    if target_variable_dv is not None:
        contingency_table_dv = pd.crosstab(target_variable_dv, explanatory_variables[column])
        chi2_dv, p_dv, _, _ = chi2_contingency(contingency_table_dv)
        chi_square_results[f"{column}_DV"] = {"Chi2 Value": chi2_dv, "P-Value": p_dv}

# Convert Chi-square results to DataFrame
chi_square_summary = pd.DataFrame(chi_square_results).transpose()
print("\nChi-square Test Results:")
print(chi_square_summary)

# Visualize Chi-square Test Results
chi_square_summary['Chi2 Value'].plot(kind='bar', figsize=(15, 6))
plt.title('Chi-square Test Results')
plt.xticks(rotation=45)
plt.ylabel('Chi2 Value')
plt.show()
```

WHATS DONE AND ACHIEVED BELOW
13. Feature Selection Using Chi-Square
Action: Use the chi-square scores to select the most relevant features.
Purpose: Identifies features that have the most significant relationships with the target variable.
Importance: Reduces dimensionality and focuses analysis on key variables that contribute meaningfully.


```python
# Feature Selection (Chi-Square)
print("\nFeature Selection using Chi-Square:")
if target_variable_dv is not None:
    chi_scores, p_values = chi2(explanatory_variables.select_dtypes(include=[np.number]), target_variable_dv)
    chi_feature_selection = pd.DataFrame({
        'Feature': explanatory_variables.select_dtypes(include=[np.number]).columns,
        'Chi2 Score': chi_scores,
        'P-Value': p_values
    })
    print(chi_feature_selection)
    chi_feature_selection.plot(x='Feature', y='Chi2 Score', kind='bar', figsize=(15, 6))
    plt.title('Feature Selection (Chi-Square)')
    plt.xticks(rotation=45)
    plt.ylabel('Chi2 Score')
    plt.show()
```
